```python
from pymongo import MongoClient
import pandas as pd
import pymongo
```

```python
client = MongoClient('mongodb://localhost:27017/')
db = client.companies
data = db.companies.find()
data_companies = pd.DataFrame(data)
```

```python
data_companies.shape
```

```python
data_companies.isna().sum().head()
```

```python
at_least_1_office = db.companies.find({
    "offices":{
        "$not":{
            "$size":0
        }
    }
})
```

```python
data_companies_1office = pd.DataFrame(at_least_1_office)
```

```python
data_companies_1office.shape
```

```python
data_companies_1office.head()
```

```python
data_companies_1office.columns
```

```python
def fixints(df, col):
    df[col].fillna(0, inplace = True)
    df[col] = df[col].astype(int)
    return df[col]

fixints(data_companies_1office, 'founded_year').head()
```

```python
data_cols = data_companies_1office[['alias_list', 'category_code','description', 'total_money_raised','founded_year','offices', ]]
display(data_cols.head())
```

```python
def latlong(df):
    df = df['offices']
    for l in df:
        if l['latitude'] and l['longitude']:
            return {
                    "lat": l['latitude'],
                    "long": l['longitude']
                }

first_office = data_cols[['offices']].apply(latlong, result_type = "expand", axis=1)
display(first_office.head())
```

```python
data_latlong = pd.concat([data_cols, first_office],axis=1)
display(data_latlong.head())
data_latlong.shape
```

```python
data_latlong.dropna(subset=['lat','long'], inplace = True)
```
